### ** 02 리스트 변동성에 근거한 변동 Sharp Ratio 계산**
1. Measures of risk-adjusted return based on volatility Sharpe ratio
1. Sharpe ratio = (Mean return − Risk-free rate) / Standard deviation of return

```python
# Sharpe Ratio 
import numpy as np
def sharpe(returns, rf, days=252):
    volatility   = returns.std() * np.sqrt(days) 
    sharpe_ratio = (returns.mean() - rf) / volatility
    return sharpe_ratio
```

```python
# 정보비율Information ratio (IR)
# 초과수익률과 적극적인 활동에 따른 수익률의 표준편차의 비율
# 이 비율이 높을수록 더 좋은 투자활동으로 판단된다
# http://www.funddoctor.co.kr/board/aboard/lecture_method_view.jsp?number=18

import numpy as np
def information_ratio(returns, benchmark_returns, days=252):
    return_difference = returns - benchmark_returns 
    volatility        = return_difference.std() * np.sqrt(days) 
    information_ratio = return_difference.mean() / volatility
    return information_ratio
```

```python
# Modigliani Ratio
import numpy as np
def modigliani_ratio(returns, benchmark_returns, rf, days=252):
    volatility   = returns.std() * np.sqrt(days) 
    sharpe_ratio = (returns.mean() - rf) / volatility 
    benchmark_volatility = benchmark_returns.std() * np.sqrt(days)
    m2_ratio     = (sharpe_ratio * benchmark_volatility) + rf
    return m2_ratio
```

## ** 3 시계열 데이터 안정성 분석 (Function for testing stationarity) **
<br>
### ** 01 ADF 검정 ( Augmented Dickey-Fuller )**
1. 해당확률변수가 **안정적인지 (시차에 의존적)** 를 판단하는 Test
1. **t -value (검정통계)** 값이, **Dickey-Fuller 임계치 (주어진 표의 값)보다 작으면 안정적**, 크면 안정적이지 못하다 
1. pandas 0.22 의 경고문구를 출력 https://github.com/statsmodels/statsmodels/issues/3814
1. Python [데이터 사이언스 코드](https://datascienceschool.net/view-notebook/ebb638fc880145b9adeef8dfa630f067/)
1. 이론설명 [link](http://blog.naver.com/PostView.nhn?blogId=jahyone20&logNo=220934340547&parentCategoryNo=&categoryNo=16&viewDate=&isShowPopularPosts=true&from=search)

```python
# 정규성 검정코드
def test_stationarity(timeseries):    

    import numpy as np
    np.warnings.filterwarnings("ignore")
    from statsmodels.tsa.stattools import adfuller

    # 시계열 데이터 이동평균 계산
    rolmean = timeseries.rolling(window=20).mean()
    rolstd  = timeseries.rolling(window=20).std()

    # 시계열 데이터 시각화
    orig = plt.plot( timeseries, color = 'blue',  label = 'Original' )
    mean = plt.plot( rolmean,    color = 'red',   label = 'Rolling Mean' )
    std  = plt.plot( rolstd,     color = 'black', label = 'Rolling Std' )
    plt.legend( loc = 'best' )
    plt.title( '이동평균(Rolling Mean) & 표준편차(Standard Deviation)' )
    plt.show( block = False )
    
    # Dickey-Fuller 테스트를 실행
    print ( 'Dickey-Fuller 테스트 결과 :' )
    dftest   = adfuller(timeseries, autolag = 'AIC')
    dfoutput = pd.Series(dftest[ 0 : 4 ], index = ['open','high','10_day_volatility', '50_day_moving_avg'])
    for  key, value  in  dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
from statsmodels.tsa.stattools import adfuller
adfuller(ts, autolag = 'AIC')
```

```python
test_stationarity(ts)
```

### ** 02 데이터를 보다 안정적인 형태로 변환하기**
**Making Data Stationary (Estimating & Eliminating Trend)**
1. log 변환
1. 이동평균선 추출하기

```python
# 정규성을 높이기 위해서 
# 데이터를 log 변환
ts_log = np.log(ts)
plt.plot(ts_log)
```

```python
# 곡선의 편차를 줄이기 위해서
# 이동평균선 추출하기
moving_avg = ts_log.rolling(10, min_periods = 1).mean()
plt.plot(ts_log)
plt.plot(moving_avg, color='red')
```

```python
# Log 변환 데이터의 차분(Diffrance) 시각화
ts_log_moving_avg_diff = ts_log - moving_avg
ts_log_moving_avg_diff.dropna( inplace = True )
test_stationarity( ts_log_moving_avg_diff )
```

<br>
### ** 03 지수 가중 이동평균 분석 (Exponentially Weighted Moving Average) **
**pandas.ewma()** : 현재에 가까운 값에 가중치를 두고서 재귀적 형태로 계산 [link](http://m.blog.daum.net/laviedor/16931708)

```python
# 지수가중 이동평균선 계산하기
expwighted_avg = pd.ewma(ts_log, halflife=12)
plt.plot(ts_log)
plt.plot(expwighted_avg, color = 'red') # expwighted_avg.plot(style = 'k--') # -- 점선을 덧칠하기
```

```python
# log 변화율과 가중이동평균의 차분(Diffrance) 확인하기
ts_log_ewma_diff = ts_log - expwighted_avg
test_stationarity( ts_log_ewma_diff )
```

<br>
### ** 04 Eliminating Trend and Seasonality**
Differencing

```python
# log 변환데이터 편차계산하기
ts_log_diff = ts_log - ts_log.shift()
plt.plot(ts_log_diff)
```

```python
ts_log_diff.dropna( inplace = True )
test_stationarity(ts_log_diff)
```

<br>
### ** 04 log변환 데이터셋을 Seasonality, Trend, random 로 분해 후 시각화**
Decomposition : 계절, 주기별 반복패턴 분석하기

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts_log, freq=52)  # 전체 52주를 단위로 나눠서 분석

trend    = decomposition.trend     # 전체적인 추세선
seasonal = decomposition.seasonal  # 계절별 반복되는 패턴분석
residual = decomposition.resid     # 잔차

plt.subplot(411); plt.plot(ts_log,   label='Original');    plt.legend(loc='best')
plt.subplot(412); plt.plot(trend,    label='Trend');       plt.legend(loc='best')
plt.subplot(413); plt.plot(seasonal, label='Seasonality'); plt.legend(loc='best')
plt.subplot(414); plt.plot(residual, label='Residuals');   plt.legend(loc='best')
plt.tight_layout()
```

```python
# 잔차값 분석
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
test_stationarity(ts_log_decompose)
```

## ** 4 ARIMA 모형을 이용한 시계열 데이터 예측하기**
**Auto Regressive Integrated Moving Average**
1. 정규화된 데이터를 대상으로
1. 관측값의 오차항 패턴을 분석
1. 같은 패턴이 지속적 적용된다는 가정하에서 예측을 한다

<br>
### ** 01. ACF & PACF Plots**
https://byeongkijeong.github.io/ARIMA-with-Python/
1. **ARMA(자기회귀 이동평균)** 모형은 AR모델과 MA모델의 결합이다
1. **AR의 특성** 을 띄는 경우, ACF는 천천히 감소하고 **PACF는 처음 시차를 제외하고 급격히 감소** 한다.
1. **MA의 특성** 을 띄는 경우, **ACF는 급격히 감소**하고 PACF는 천천히 감소

```python
# AR Model: ACF and PACF plots:

# from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
# plot_acf(ts_log_diff)
# plot_pacf(ts_log_diff)

from statsmodels.tsa.stattools import acf, pacf  

lag_acf  = acf( ts_log_diff, nlags=20)
lag_pacf = pacf(ts_log_diff, nlags=20, method='ols')

plt.figure(figsize=(8,4))
plt.plot(lag_acf,  label='ACF')   # Plot ACF:    
plt.plot(lag_pacf, label='PACF')  # Plot PACF:
plt.axhline(y = 0, linestyle = '--', color = 'gray')
plt.axhline(y = -1.96 / np.sqrt(len(ts_log_diff)), linestyle = '--', color = 'gray')
plt.axhline(y =  1.96 / np.sqrt(len(ts_log_diff)), linestyle = '--', color = 'gray')
plt.title('Partial Autocorrelation Function'); plt.tight_layout(); plt.legend()
```

<br>
### ** 02. ARIMA 모델링**
https://byeongkijeong.github.io/ARIMA-with-Python/
1. **_p_** : **AR모형(자기회귀모형)** 의 Lag(시차) 값으로, p는 실증적으로 결정한다 
1. **_q_** : **MA모형(이동평균모형)** 의 Lag(시차) 값으로, q는 실증적으로 결정한다
1. **ARMA(자기회귀 이동평균)** 모형은 AR모델과 MA모델의 결합으로, 차분(Diffrence)횟수를 의미하는 **_d_** 를 추가로 필요로 한다
1. ARIMA의 모수는 통상적으로 ** p + q < 2 ** 그리고  **p x q = 0** 인 값을 많이 사용하는데, 이는 AR/ MA 모델중 하나의 경향을 강하게 띔을 전제로 분석함이 효과적이다
1. **p 와 d, q의 추정**은, **ACF(Autocorrelation function) plot** 와 **PACF(Partial autocorrelation function) plot** 을 사용한다

```python
# results_AR : AR model Result
from statsmodels.tsa.arima_model import ARIMA

model      = ARIMA(ts_log, order=(2, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ts_log_diff)**2))
```

```python
# results_MA : MA Model Result
model      = ARIMA(ts_log, order=(0, 1, 2))  
results_MA = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ts_log_diff)**2))
```

```python
# results_ARIMA : ARIMA Model Result
model = ARIMA(ts_log, order=(2, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS (Root Squared Sum): %.4f'% sum((results_ARIMA.fittedvalues-ts_log_diff)**2))
```

<br><br>
## ** 5 데이터를 원본 형태로 되돌리기**
**Convert to original scale:**

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff.head()
```

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_diff_cumsum.head()
```

```python
predictions_ARIMA_log = pd.Series(ts_log.iloc[0], index=ts_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum, fill_value=0)
predictions_ARIMA_log.head()
```

```python
# ARIMA로 예측한 Log 변환율 시각화
plt.plot(ts_log)
plt.plot(predictions_ARIMA_log)
```

```python
# ARIMA로 예측한 원본 데이터 시각화
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(ts)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-ts)**2)/len(ts)))
```

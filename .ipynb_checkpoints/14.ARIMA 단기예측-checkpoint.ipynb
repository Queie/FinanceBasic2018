## **1. Loading Stock Data into Pandas**
<br>
### ** 01 주가 데이터 불러오기**
googlefinance.get

```python
%matplotlib inline
from googlefinance.get import get_data
data = get_data("KRX:030200", '2M')
series = data.close
series.plot(figsize=(9,4))
data.info()
```

<br>
### ** 01 주가 데이터 불러오기**
googlefinance.get

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

series = data.close
plot_acf(series)
plot_pacf(series)
plt.show

# ACF : 5 Time Lag 까지 급격히 떨어진 뒤, 요동을 친다
# PACF : 1에서 .8을 찎은 뒤 급격히 떨어진다
```

```python
# 1차 차분(Diff)한 뒤 변화를 살펴본다
diff_1 = series.diff(periods=1).iloc[1:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
# 차분 이후의 ACF, PACF를 본 결과 PACF 경향이 더 도드라 진다
# PCAF가 안전된 만큼, MA 성향이 더 도드라 진다
```

## **2. ARIMA 모델로 주가 예측하기**
<br>
### ** 01 ARIMA 모형 구축**
1차 차분의 시계열이, 정상상태(Stationary)를 보이므로, 데이터에는 ARIMA(0,1,1)을 사용하기로 한다.

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(series, order = (0, 1, 1))
model_fit = model.fit(trend = 'c', full_output = True, disp = 1)
print(model_fit.summary())
# P > |z| 값이 1차차분결과 0.5보다 작으므로 정규성에 약간 부족하다
```

```python
model = ARIMA(series, order = (0,1,3))
model_fit = model.fit(trend = 'nc', full_output = True, disp = 1)
print(model_fit.summary())
```

```python
# 위에서 생성한 모델을 사용하여 주가를 예측한다
model_fit.plot_predict()
```

```python
#  마지막 예측값을 수치화 한다
fore = model_fit.forecast(steps = 5)
print('주가예측 : {} \nstderr : {} \nupper bound : {} \nlower bound : {}'.format(
    fore[0], fore[1][0], fore[2][0][0], fore[2][0][1]))
pd.Series(fore[0]).plot(figsize=(6,2))
```

### **01 총 수익률**
1. 1년에 6% 인상을 (결정론으로) 가정한다면, 원금의 2배가 되는 기간은?
1. 분석결과 : 2022년에 2배를 넘어선다

```python
import matplotlib.pyplot as plt
principle_value  = 10000  # 초기 투자비용
grossReturn      = 1.06   # 투자이익률
year             = 2010   # 투자시점
return_amt, x, y = [], [], [10000]
return_amt.append(principle_value)
x.append(year)

for i in range(1,15):
    return_amt.append(return_amt[i-1] * grossReturn)
    print ("Year-{:2} Returned:{}".format(i, round(return_amt[i], 3)))
    year += 1
    x.append(year)
    y.append(833.33 * (year-2010) + principle_value)

# Visualization
plt.grid()
plt.plot(x,return_amt, color='r')
plt.plot(x,y, color='b')
```

```python
def printHeaders(term, extra):

    # Print headers
    print ("""\nExtra-Payment: $  {}  Term: {} + years
           \n---------------------------------------------------------""".format(str(extra), str(term)))
#     'Pmt no'.rjust(6), ' ', 'Beg. bal.'.ljust(13), ' ',
#     'Payment'.ljust(9), ' ', 'Principal'.ljust(9), ' ',
#     'Interest'.ljust(9), ' ', 'End. bal.'.ljust(13)
#     ''.rjust(6, '-'), ' ', ''.ljust(13, '-'), ' ',
#     ''.rjust(9, '-'), ' ', ''.ljust(9, '-'), ' ',
#     ''.rjust(9, '-'), ' ', ''.ljust(13, '-'), ' '
```

```python
def amortization_table(principal, rate, term, extrapayment, printData=False):
    xarr = []
    begarr = []
    original_loan = principal
    money_saved=0
    total_payment=0
    payment = pmt(principal, rate, term)
    begBal = principal
    # Print data
    num=1
    endBal=1
    if printData == True: 
        printHeaders(term, extrapayment)

    while (num < term + 1) and (endBal >0):
        interest = round(begBal * (rate / (12 * 100.0)), 2)
        applied  = extrapayment + round((payment - interest), 2)
        endBal   = round(begBal - applied, 2)
        if (num-1)%12 == 0   or   (endBal < applied+extrapayment):
            begarr.append(begBal)
            xarr.append(num/12)
            
            if printData == True:
                print ('{0:3d}'.format(num).center(6), 
                       '\n {0:,.2f}'.format(begBal).rjust(13),
                       '\n {0:,.2f}'.format(payment).rjust(9),
                       '\n {0:,.2f}'.format(applied).rjust(9),
                       '\n {0:,.2f}'.format(interest).rjust(9),
                       '\n {0:,.2f}'.format(endBal).rjust(13))
            total_payment += applied+extrapayment
            num +=1
            begBal = endBal

        if extrapayment > 0 :
            money_saved = abs(original_loan - total_payment)
            print ('\nTotal Payment:','{0:,.2f}'.format(total_payment).rjust(13),
                   '\nMoney Saved:','{0:,.2f}'.format(money_saved).rjust(13))
        return xarr, begarr, '{0:,.2f}'.format(money_saved)
```

```python
# Convert to decimal and round off to two decimal
def pmt(principal, rate, term):
    ratePerTwelve = rate / (12 * 100.0)
    result        = principal * (ratePerTwelve / (1 - (1 + ratePerTwelve) ** (-term)))
    #result        = Decimal(result)
    result        = round(result, 2)
    return result
```

```python
#amortization_table(150000, 4, 180, 500)
i          = 0
markers    = ['o','s','D','^','v','*','p','s','D','o','s','D','^','v','*','p','s','D']
markersize = [8, 8, 8, 12, 8, 8, 8, 12, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8]
for extra in range(100,1700,100):
    xv, bv, saved = amortization_table(450000, 5, 360, extra, False)
    if extra == 0:
        plt.plot(xv, bv, 
                 color = colors[i], 
                 lw    = 2.2, 
                 label = 'Principal only',
                 marker = markers[i], 
                 markersize = markersize[i])
    else:
        plt.plot(xv, bv, 
                 color  = colors[i], 
                 lw     = 2.2, 
                 label  = "Principal plus\$"+str(extra)+str("/month, Saved:\$")+saved, 
                 marker = markers[i],
                 markersize=markersize[i])
        i +=1

plt.figure(figsize=(18, 14))
plt.xlabel('Years', fontsize=18)
plt.ylabel('Mortgage Balance', fontsize=18)
plt.title("Mortgage Loan For $350,000 With Additional Payment Chart", fontsize=20)
plt.grid(True); plt.legend(); plt.show()
```

### **02 재고문제에  몬테카를로 시뮬레이션 적용**

1. 1개팔면 60센트가 이익, 재고로 남으면 40센트 손실
1. Demand(수요) =  80 ~ 140 범위내에서 일정하게 분포한다
1. 이때 수익으 최대로 하는 주문량을 예측해보자

```python
import numpy as np
from math import log
import matplotlib.pyplot as plt

# 수익을 계산하는 함수
def generate_Profit(d):
    global s
    if d >= s: return 0.6*s             # 물건이 팔린경우
    else:      return 0.6*d - 0.4*(s-d) # 재고로 남은경우

x, y, maxprofit = [], [], 0

# 몬테카를로 시뮬레이션 적용
for s in range (20, 305):
    for i in range(1,1000):  # 1,000 : 실험횟수 (10,000번해도 결과는 비슷) 
        d      = np.random.randint(10, high=200)  # 랜덤수 생성
        profit = generate_Profit(d)               # 수익을 계산
        if profit > maxprofit:  maxprofit = profit
        x.append(s)               # x축 단위값 저장
        y.append(log(maxprofit))  # 수익률을 Log 스케일로 저장

plt.plot(x,y)
print ("Max Profit:",maxprofit)
```

### **03 생활 속 문제를  몬테카를로 시뮬레이션 적용**
30명이 모인 집단에서 생일이 같을 확률의 계산

```python
import numpy as np
student, Trials, SameBirth = 30, 10000, 0

for trial in range(Trials):
    
    year = [0] * 365                      # [0,0,0,...] : 0을 365개 갖는 list
    for i in range(student):
        new_Day = np.random.randint(365)  # 임의의 생일난수 30개를 생성
        year[new_Day] = year[new_Day] + 1 # 1을 더한다 ('0'을 계산에서 방지)
        have_SameBday = False             # 값 초기화

    for num in year:    # 전체 365개의 난수를 돌면서 
        if num > 1:     # 30개의 입력값 위치를 찾는다
            have_SameBday = True

    if have_SameBday == True:
        SameBirth += 1
        
    if trial % 999 == 0:
        prob = float(SameBirth) / float(Trials)
        print("{:4}\'th test {} 명중 2명의 생일이 같을확률 : {}".format(trial, student, prob))
```

```python
# 2점 시도하는 경우
def attemptTwo():
    havePossession, pointsDown, timeLeft = True, 3, 30
    while (timeLeft > 0):     
        if (havePossession):           # 1. 점유권을 갖는 경우
            if (pointsDown >= 3):      # 3점 이상을 뒤진다면, 2점을 빠르게 득점
                timeLeft -= timeToShoot2
            else:                      # 2점 이하로 뒤진다면 먼저 시간을 끌어야 한다
                timeLeft = 0

            if (np.random.randint(0, high=100) < twoPtPercent): # 슛 결정 후, 성공률 계산
                pointsDown -= 2
                havePossession = False
            else:
                if (np.random.randint(0, high=100) >= offenseReboundPercent):  # 상대방이 리바운드 하는 경우
                    havePossession = False                                     # 점유권을 잃는다
        else:                           # 2. 점유권을 잃은 경우
            if (pointsDown > 0):        # 점유권을 찾기위해 파울을 해야 한다
                timeLeft -= timeToFoul  # 파울을 위한 시간
                if (np.random.randint(0, high=100) < oppFtPercent):   # 상대방이 프이드로우 획득시
                    pointsDown += 1
                if (np.random.randint(0, high=100) < oppFtPercent): 
                    pointsDown += 1
                    havePossession = True

            else:   # 놓친 프리드로우를 리바운드 한다
                if (np.random.randint(0, high=100) >= ftReboundPercent):
                    havePossession = True
                else: # 동점 또는 이기고 있는 경우
                    if (np.random.randint(0, high=100) < oppTwoPtPercent):
                        pointsDown += 2  # 상대팀이 2점을 획득하는 경우
                        timeLeft = 0
                    if (pointsDown > 0):
                        return False
                    else:
                        if (pointsDown < 0):
                            return True
                        else:
                            if (np.random.randint(0, high=100) < overtimePercent):
                                return True
                            else:
                                return False
```

```python
for i in range(5):                 # can be run individually as well
    x, y1, y2 = [], [], []
    threePtPercent, twoPtPercent  = threePercents[i], twoPercents[i]  # 3점슛 기회, 2점슛 기회 %
    oppTwoPtPercent, oppFtPercent = 40, 70      # 상대팀 2점슛 기회 %, 상대팀 프리드로우 %
    timeToShoot2, timeToFoul    = 5, 5          # 2점슛을 위해 경과한 시간, 상대파울로 경과한 시간
    offenseReboundPercent = 25                  # 공격중 리바운드 %
    ftReboundPercent, overtimePercent = 15, 50  # FT를 놓친 공격의 리바운드 %, 연장경기에서 승리할 확률 %
    winsTakingThree, lossTakingThree, winsTakingTwo, lossTakingTwo = 0, 0, 0, 0
    trials, curTrial = 400, 1
    while curTrial < trials:
        if (attemptThree()): 
            winsTakingThree += 1  # 3점슛 시도
        else:                
            lossTakingThree += 1
            if attemptTwo() == True : # 2점슛 시도
                winsTakingTwo += 1
            else:
                lossTakingTwo += 1
            x.append(curTrial)
            y1.append(winsTakingThree)
            y2.append(winsTakingTwo)
            curTrial += 1
```

```python
plt.plot(x,y1, color=colors[colind], label=names[i]+" Wins Taking Three Point", linewidth=2)
plt.plot(x,y2, color=colors[20], label=names[i]+" Wins Taking Two Point", linewidth=1.2)
colind += 2
legend = plt.legend(loc='upper left', shadow=True,)
for legobj in legend.legendHandles:
    legobj.set_linewidth(2.6)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

rect = [0.1, 5.0, 0.1, 0.1]
sigma, S0 = 0.04, 20        #  
T, mu, dt = 2, 0.1, 0.01
N = round( T / dt )
t = np.linspace(0, T, N)

# Standare normal distrib
W = np.random.standard_normal(size = N)
W = np.cumsum(W) * np.sqrt(dt)
X = (mu - 0.5 * sigma ** 2) * t + sigma * W

S = S0 * np.exp(X)   # 브라운 운동 모델
plt.figure(figsize = (8,4))
plt.plot(t, S, lw = 2)
plt.xlabel("Time t", fontsize=16)
plt.title("Geometric Brownian Motion (Simulation)", fontsize = 18)
plt.show()
```

주가데이터를 활용한 브라운 모델 시뮬레이션

```python
import pylab, random
class Stock(object):
    
    def __init__(self, price, distribution):
        self.price        = price
        self.history      = [price]
        self.distribution = distribution
        self.lastChange   = 0

    def setPrice(self, price):
        self.price = price
        self.history.append(price)

    def getPrice(self):
        return self.price
    
    def walkIt(self, marketBias, mo):
        oldPrice   = self.price
        baseMove   = self.distribution() + marketBias
        self.price = self.price * (1.0 + baseMove)
        if mo:
            self.price = self.price + random.gauss(.5, .5)*self.lastChange
        if self.price < 0.01:
            self.price = 0.0
        self.history.append(self.price)
        self.lastChange = oldPrice - self.price

    def plotIt(self, figNum):
        pylab.figure(figNum)
        pylab.plot(self.history)
        pylab.title('Closing Price Simulation Run-' + str(figNum))
        pylab.xlabel('Day')
        pylab.ylabel('Price')
```

```python
def testStockSimulation():
    def runSimulation(stocks, fig, mo):
        mean = 0.0
        for s in stocks:
            for d in range(numDays):
                s.walkIt(bias, mo)
            s.plotIt(fig)
        mean += s.getPrice()
        mean = mean/float(numStocks)
        pylab.axhline(mean)
        pylab.figure(figsize=(12,12))
        numStocks = 20
        numDays = 400
        stocks = []
        bias = 0.0
        mo = False
        startvalues = [100,500,200,300,100,100,100,200,200, 300,300,400,500,100,300,100,100,100,200,200,300]
        for i in range(numStocks):
            volatility = random.uniform(0,0.2)
            d1 = lambda: random.uniform(-volatility, volatility)
            stocks.append(Stock(startvalues[i], d1))
        runSimulation(stocks, 1, mo)
```

### **02 확산 기반 시뮬레이션**
보다 상세한 모델 시뮬레이션이 가능하다

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as npr
S0 = 100 # initial value
sigma, theta = 0.25, 0.24
r, T, x0, k = 0.05, 2.0, 0, 1.8
i, M = 100000, 50
dt = T / M
```

```python
def srd_euler():
    xh = np.zeros((M + 1, i))
    x1 = np.zeros_like(xh)
    xh[0] = x0
    x1[0] = x0
    for t in range(1, M + 1):
        xh[t] = (xh[t - 1]
                 + k * (theta - np.maximum(xh[t - 1], 0)) * dt
                 + sigma * np.sqrt(np.maximum(xh[t - 1], 0)) * np.sqrt(dt)
                 * npr.standard_normal(i))
        x1 = np.maximum(xh, 0)
    return x1
```

```python
x1 = srd_euler()
plt.figure(figsize=(6,5))
plt.hist(x1[-1], bins=30, color='#98DE2f', alpha=0.85)
plt.xlabel('value')
plt.ylabel('frequency')
plt.grid()
```

### **01 셀링의 분리 모델**
보다 상세한 모델 시뮬레이션이 가능하다

\section{Black-Scholes Equation}
The Black-Scholes equation describes the price of an option over time. The derivation of this equation is complex and exceeds the scope of this paper, 
so we simply provide the equation.

\begin{equation}
	\frac{\partial \mathrm C}{ \partial \mathrm t } + \frac{1}{2}\sigma^{2} \mathrm S^{2} \frac{\partial^{2} \mathrm C}{\partial \mathrm C^2}
	+ \mathrm r \mathrm S \frac{\partial \mathrm C}{\partial \mathrm S}\ =
	\mathrm r \mathrm C 
	\label{eq:1}
\end{equation}

\section{Black-Scholes formula for European option price}
The Black-Scholes formula allows us the calculate the price of European call and put options.
\begin{equation}
	\mathrm C(\mathrm S,\mathrm t)= \mathrm N(\mathrm d_1)\mathrm S - \mathrm N(\mathrm d_2) \mathrm K \mathrm e^{-rt}
	\label{eq:2}
\end{equation}


```python
from math import log, sqrt, exp
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
colors = [(31, 119, 180),  (174, 199, 232), (255,128,0),     (255, 15, 14),   (44, 160, 44), 
          (152, 223, 138), (214, 39, 40),   (255, 152, 150), (148, 103, 189), (197, 176, 213),
          (140, 86, 75),   (196, 156, 148), (227, 119, 194), (247, 182, 210), (127, 127, 127),
          (199, 199, 199), (188, 189, 34),  (219, 219, 141), (23, 190, 207), (158, 218, 229)]

# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.
for i in range(len(colors)):
    r, g, b   = colors[i]
    colors[i] = (r / 255., g / 255., b / 255.)
```

```python
def black_scholes_merton(S, r, sigma, X, T):
    S               = float(S) # convert to float
    logsoverx       = log (S/X)
    halfsigmasquare = 0.5 * sigma ** 2
    sigmasqrtT      = sigma * sqrt(T)
    d1              = logsoverx + ((r + halfsigmasquare) * T) / sigmasqrtT
    d2              = logsoverx + ((r - halfsigmasquare) * T) / sigmasqrtT
    # stats.norm.cdf() : 누적분포 함수
    value           = (S * stats.norm.cdf(d1, .0, 1.) - X * exp(-r * T) * stats.norm.cdf(d2, 0.0, 1.0))
    return value
```

```python
def vega(S, r, sigma, X, T):
    S               = float(S)
    logsoverx       = log (S/X)
    halfsigmasquare = 0.5 * sigma ** 2
    sigmasqrtT      = sigma * sqrt(T)
    d1              = logsoverx + ((r + halfsigmasquare) * T) / sigmasqrtT
    vega            = S * stats.norm.cdf(d1, 0.0, 1.0) * sqrt(T)
    return vega
```

```python
def impliedVolatility(S, r, sigma_est, X, T, Cstar, it):
    for i in range(it):
        numer      = (black_scholes_merton(S, r, sigma_est, X, T) - Cstar)
        denom      = vega(S,r, sigma_est, X, T)
        sigma_est -= numer/denom
    return sigma_est
```

```python
import pandas as pd
h5           = pd.HDFStore('./data/vstoxx_data_31032014.h5', 'r')
futures_data = h5['futures_data'] # VSTOXX futures data
options_data = h5['options_data'] # VSTOXX call option data
h5.close()

# V0 : 인덱스 종가 , r : 무위험 이자율, tol : 포트폴리오 전체비율
V0, r, sigma_est, tol   = 17.6639, 0.04, 2, 0.5
options_data['IMP_VOL'] = 0.0

for option in options_data.index:
    futureval = futures_data[futures_data['MATURITY'] == options_data.loc[option]['MATURITY']]['PRICE'].values[0]
    # 적당한 예측값 선택하기
    if (futureval * (1 - tol) < options_data.loc[option]['STRIKE']< futureval * (1 + tol)):
        impliedVol = impliedVolatility(V0,r,sigma_est,
                                       options_data.loc[option]['STRIKE'],
                                       options_data.loc[option]['TTM'],
                                       options_data.loc[option]['PRICE'], it=100)
        options_data['IMP_VOL'].loc[option] = impliedVol
```

```python
plot_data  = options_data[options_data['IMP_VOL'] > 0]
maturities = sorted(set(options_data['MATURITY']))
i          = 0
plt.figure(figsize=(15, 10))

for maturity in maturities:
    data      = plot_data[options_data.MATURITY == maturity]
    plot_args = {'lw':3, 'markersize': 9}
    plt.plot(data['STRIKE'], data['IMP_VOL'], 
             label  = maturity.date(),
             marker = 'o', 
             color  = colors[i], **plot_args)
    i += 1
plt.grid(True)
plt.xlabel('Strike rate $X$', fontsize=18)
plt.ylabel(r'Implied volatility of $\sigma$', fontsize=18)
plt.title('Short Maturity Window (Volatility Smile)', fontsize=22)
plt.legend()
plt.show()
```

```python
# 1 - 2 모듈을 호출 (보조 프로그램 실행)  
import pandas as pd

# 1 - 1 'df' 객체에 추출한 정보를 입력 #
# .read_html : 메서드 (입력값 필요)
df  = pd.read_html("http://media.kisline.com/highlight/mainHighlight.nice?nav=1&paper_stock=000660")[6]  
df  = df.replace('-',0)
df  = df.set_index(df.columns[0])

# .column : 메서드 (입력값 불필요)
for col in df.columns:
    df[col] = [float(i) for i in df[col]]  #df[col].apply(lambda x: float(x))
df.head()
```

### **05 사용자 함수 만들기**
1. 위에서 작업한 내용들을 1개의 함수로 묶는다
1. 즉 하청기업에 떠넘김으로써, 중간재 없이 최종 결과만 챙길 수 있다
1. 이로써 메모리 및 리소스 를 아낄 수 있다

```python
def finance_table(code):
    import numpy as np   
    import pandas as pd

    df  = pd.read_html("http://media.kisline.com/highlight/mainHighlight.nice?nav=1&paper_stock=" + code)[6]  
    df  = df.replace('-',0)
    df  = df.set_index(df.columns[0])

    for col in df.columns:
        df[col] = df[col].apply(lambda x: float(x))
    df.iloc[0,:].plot()
    df.iloc[1,:].plot()
    df.iloc[3,:].plot()
    return df
```

```python
finance_table('005930')
```

```python
# 3. 함수로 만들기 
def total_table():
    import pandas as pd
    url = "http://finance.daum.net/quote/all.daum?nil_stock=refresh"
    toplist = pd.read_html(url)
    result = [table  for i, table in enumerate(toplist)   if i >0]
    df = pd.concat(result, axis=0).dropna()
    df.columns  = df.iloc[0,:]
    df = df.drop(df.index[0])
    return df

total_table()
```
